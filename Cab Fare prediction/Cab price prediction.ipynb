### import modules

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from sklearn.tree import DecisionTreeRegressor
import lightgbm as lgbm

pd.set_option('display.max_colwidth', None)

import warnings
warnings.simplefilter('ignore')
```

```python
def get_rmse(model, X_train, X_test, y_train, y_test):
    _ = model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return np.sqrt(mean_squared_error(y_test, preds))
```

### Feature engineering

```python
X = train.drop('fare', axis=1)
y = train.fare
```

```python
from feature_engine.encoding import MeanEncoder()
'''The MeanEncoder() replaces categories by the mean value of the target for eachcategory.'''

encoder = MeanEncoder(variables=cat_features)

encoder.fit(X,y)

X = encoder.transform(X)
```

```python
# data splitting
X_train, X_test,  y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

# Modelling

```python
base_model = DecisionTreeRegressor(random_state=1)
get_rmse(base_model,X_train, X_test,  y_train, y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV

hyperparam_combs = {
    'max_depth': [4, 6, 8, 10, 12, 14, 16],
    'min_samples_split': [2, 10, 20, 30, 40],
    'max_features': [0.2, 0.4, 0.6, 0.8, 1],
    'max_leaf_nodes': [8, 16, 32, 64, 128,256]
}

search = RandomizedSearchCV(DecisionTreeRegressor(random_state=1),
                         hyperparam_combs,
                         scoring='neg_mean_squared_error',
                         random_state=1,
                         cv=3,
                         n_iter=20)

search.fit(X_train, y_train)

search.best_params_
```

```python
get_rmse(search.best_estimator_,X_train, X_test,  y_train, y_test)
```

```python
lgbm_model = lgbm.LGBMRegressor(random_state=1)
get_rmse(lgbm_model,X_train, X_test,  y_train, y_test)
```

```python
lgbm_params = dict(
    boosting_type=['gbdt','goss'],
    num_leaves=[20,30,40,50],
    max_depth=[-1, 5, 10, 15],
    learning_rate=[0.05,0.07,0.1],
    n_estimators=[100,500,1000],
    subsample=[0.25, 0.50,1.0]
)
lgbm_model = lgbm.LGBMRegressor(objective='regression',random_state=1)
```
